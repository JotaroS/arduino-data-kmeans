# BID data analysis

Run gather_data to spit out `out.csv`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans

%matplotlib inline
```

```python
df = pd.read_csv('out.csv')
```

```python
df.describe()
```

## visualization

the data roughly looks like having three mountains, but the two peak is somewhat confusing.

```python
sns.distplot(df.data)
```

```python
sns.lineplot(data=df, x='count', y='data')
```

## analysis

run sklearn k-means clustering

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(df.data.values.reshape(-1,1))
pred = kmeans.predict(df.data.values.reshape(-1,1))
df['pred'] = pred
```

```python
sns.scatterplot(data=df, x='count',y='data',hue='pred')
```

```python
sns.catplot(data=df,x='pred',y='data',kind="violin")
```

```python
sns.distplot(df.data, hue='pred')
```
